```python
import numpy as np
import matplotlib.pyplot as plt

# I uniformly generated 50 samples from 0 to 10
x = np.random.uniform(0, 10, 50)

# Generated epsilon from N(0, 2)
epsilon = np.random.normal(0, 2, 50)

y = 3 * x + 5 + epsilon

n = len(x)

sum_x = np.sum(x)
sum_y = np.sum(y)
sum_x_squared = np.sum(x**2)
sum_xy = np.sum(x * y)

# Calculate the slope (m) and intercept (c) using the formulas taught

m = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x**2)
c = (sum_y * sum_x_squared - sum_x * sum_xy) / (n * sum_x_squared - sum_x**2)


y_pred = m * x + c

# Plot the sample of x and y
plt.scatter(x, y, color='blue', label='Data Points')

# Plot the manually calculated regression line
plt.plot(x, y_pred, color='green', label='Regression Line')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression of y = 3x + 5 + epsilon')
plt.legend()
plt.show()

```
