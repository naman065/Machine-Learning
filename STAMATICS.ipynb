```python
"""IN Linear Regression we have Weights and Biases and the gradient descent algorithm
is used to minimize the error by changing these weights.

This Newton-Rhapson method can be used to minimize the  error function in linear regression model
"""
def gradient_descent(X, y, learning_rate, num_iterations):
    # Initialize parameter theta (our weight matrix)
    #Use the np.random.randn function for the same.
    theta = np.random.randn("X.shape[1]")#since no. of features is one so in
    #y=theta*x+c we optmize theta

    m = len(y)
    cost_history = []

    for i in range(num_iterations):
        # Calculate predicted values:
        predictions = np.dot(X,theta)# for every X we do y=theta*x

        # Calculate the error: use the mse function
        error = predictions-y #since this is usd in the partial derivative  hence cant use mse directly here

        # Calculate the gradients: 1/m * X.T * error
        gradients = (1/m) * np.dot(X.T, error)# 1/m(theta*X-y)*theta in vect form

        # Update parameters: Apply the GD algorithm
        theta -= learning_rate * gradients# decreasing weight in direction that mimimizes cost function

        # Calculate the cost function: MSE
        cost = np.mean((error)**2)
        cost_history.append(cost)

    return theta, cost_history
```

***Advantage of L1 norm over L2 norm:

L1 norm (Manhattan Distance) helps us focus on the most important things. Imagine we have a bunch of items, some of which we rarely use. L1 norm is like telling us, "Just focus on the items we use frequently and ignore the rest." It helps us declutter and focus on what matters most.
HERE ITEM MEANS WEIGHTS IN TERMS OF LR!

Advantage of L2 norm over L1 norm:

L2 norm(Euclidian Distance) helps us find a solution that's stable and doesn't change too much. Think of it like building a sturdy table. We want the legs of the table to be evenly strong so that if we put something heavy on one side, the table doesn't wobble too much. L2 norm does this by spreading out the "strength" evenly across all parts of the solution, making it less likely to wobble or change drastically with small adjustments.
That is it distributes error over the coefecients aka the weights!
